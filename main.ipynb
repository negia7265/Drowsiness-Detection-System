```python


mixer.init()
sound=mixer.Sound('alarm.wav')
# Load Dlib's face detector (HOG-based) and the facial landmark predictor
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')  # Ensure this file is available

# model = load_model('your_model.h5') 

# Start video capture
cap = cv2.VideoCapture(0)

score=0 
margin_horizontal = 18  
margin_vertical = 35    
detected_eyes = []

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Detect faces in the frame
    faces = detector(gray)
    height,width=frame.shape[0:2]
    for face in faces:
        # Get the landmarks/parts for the face
        landmarks = predictor(gray, face)
        
        # Get coordinates for the left and right eye
        left_eye_points = [(landmarks.part(n).x, landmarks.part(n).y) for n in range(36, 42)]
        right_eye_points = [(landmarks.part(n).x, landmarks.part(n).y) for n in range(42, 48)]
        
        # Compute bounding box for left eye
        x_left_min = max(min([p[0] for p in left_eye_points]) - margin_horizontal, 0)
        y_left_min = max(min([p[1] for p in left_eye_points]) - margin_vertical, 0)
        x_left_max = min(max([p[0] for p in left_eye_points]) + margin_horizontal, frame.shape[1])
        y_left_max = min(max([p[1] for p in left_eye_points]) + margin_vertical, frame.shape[0])
        
        # Extract and preprocess left eye
        left_eye = frame[y_left_min:y_left_max, x_left_min:x_left_max]
        if(left_eye.size>0):
            left_eye_resized = cv2.resize(left_eye, (80, 80))  # Resize to (80, 80) as required by the model

            detected_eyes.append(left_eye_resized)

            left_eye_resized = left_eye_resized / 255.0  # Normalize
            left_eye_reshaped = np.reshape(left_eye_resized, (1, 80, 80, 3))
        
            cv2.rectangle(frame, (x_left_min, y_left_min), (x_left_max, y_left_max), (0, 255, 0), 2)
        
        # Compute bounding box for right eye
        x_right_min = max(min([p[0] for p in right_eye_points]) - margin_horizontal, 0)
        y_right_min = max(min([p[1] for p in right_eye_points]) - margin_vertical, 0)
        x_right_max = min(max([p[0] for p in right_eye_points]) + margin_horizontal, frame.shape[1])
        y_right_max = min(max([p[1] for p in right_eye_points]) + margin_vertical, frame.shape[0])
        
        # Extract and preprocess right eye
        right_eye = frame[y_right_min:y_right_max, x_right_min:x_right_max]
        if(right_eye.size>0):
            right_eye_resized = cv2.resize(right_eye, (80, 80))  

   

            right_eye_resized = right_eye_resized / 255.0 
            right_eye_reshaped = np.reshape(right_eye_resized, (1, 80, 80, 3))


            right_predict=model.predict(right_eye_reshaped)
    
            if(right_predict>=0.55):
                cv2.putText(frame,'open',(10,height-20),fontFace=cv2.FONT_HERSHEY_COMPLEX,fontScale=0.8,color=(255,255,255),thickness=1)
                score=score-1
                if score<0:
                    score=0
            else:
                cv2.putText(frame,'closed',(10,height-20),fontFace=cv2.FONT_HERSHEY_COMPLEX,fontScale=0.8,color=(255,255,255),thickness=1)
                score=score+1
                if score>3:
                    sound.play()
          
            cv2.rectangle(frame, (x_right_min, y_right_min), (x_right_max, y_right_max), (0, 255, 0), 2)

    # Display the resulting frame
    cv2.imshow("Eyes Detection", frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release video capture and close windows
cap.release()
cv2.destroyAllWindows()
```
